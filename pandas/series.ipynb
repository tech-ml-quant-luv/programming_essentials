```python
import pandas as pd
```

#### Series

    -   A series is a single COLUMN of data
    -   Intuitively, a pandas series is the best features of a python list and a python dictionary combined into one single object.
    -   Python list is best suited to save the data in an order and the dictionary is best suited to show associations.
    


```python
ice_cream = ["Chocolate", "Vanilla", "Strawberry", "Rum Raisin"]
pd.Series(ice_cream)
```

##### Notes

    - The series is going to maintain the original order of the data from the list.
    - Whenever pandas do not get a explicit statement from us about the label should be, it automatically gives them a numerical assignment.
    - Except simple data types like integers and booleans floats etc, pandas will always return dtypes as objects.

```python
lottery_numbers = [4,8,15,28,42]
pd.Series(lottery_numbers)
```

```python
registrations = [True, False, True, False, False, True , False , True , True]
pd.Series(registrations)
```

```python
sushi = {
    "Salmon": "Orange",
    "Tuna" : "Red", 
    "Eel" : "Brown"
}
pd.Series(sushi)    #Keys become the index values.
```

##### Series Methods

```python
prices = pd.Series(
    [
        2.99,4.45,1.36
    ]
)
prices
```

```python
prices.sum()
```

```python
prices.product()
```

```python
prices.mean()
```

```python
prices.std()    # Standard Deviation
```

##### Parameters and Arguments

    -   Parameter is a name to a expected input to a function or method. It is a placeholder.
    -   An argument is the concrete valeu we provide for a parameter during invocation

```python
fruits = ["Apple", "Orange","Plum", "Grape", "Blueberry", "Watermelon"]
weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Monday"]
```

```python
pd.Series(fruits)
pd.Series(weekdays)
pd.Series(fruits, weekdays) #The first argument becomes the data and the second argument becomes the index of that data
pd.Series(weekdays,fruits)  # index and data are interchanged, oeders matters
```

```python
pd.Series(data=fruits,index=weekdays)
```

```python
pd.Series(fruits, index=weekdays)
```

```python
pd.Series(fruits)   # This will create a numeric index
```
